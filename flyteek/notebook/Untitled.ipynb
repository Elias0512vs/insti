```python
# 1. Contar la frecuencia de cada curso.
conteo_cursos = df['Curso'].value_counts()

# 2. Filtrar el conteo para quedarte con los que tienen más de 100 alumnos.
cursos_principales = conteo_cursos[conteo_cursos > 30]
conteo_otros=conteo_cursos[conteo_cursos <= 30].sum()
# 3. Plotear el resultado.
if conteo_otros > 0:
    conteo_final = pd.concat([cursos_principales, pd.Series({'Otros': conteo_otros})])
else:
    conteo_final = cursos_principales

# Crear el gráfico
if not conteo_final.empty:
    plt.figure(figsize=(10, 6))
    conteo_final.sort_values(ascending=False).plot(kind='bar', color='purple')
    plt.title('Distribución de Alumnos por Curso (Consolidado mayores que 30) ')
    plt.xlabel('Curso')
    plt.ylabel('Número de Alumnos')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
else:
    print("No hay datos de alumnos para graficar.")
```

```python
veces_llevado=dp["Correo"].value_counts()
retencion=veces_llevado[veces_llevado>1]
alumnos_con_mas_de_un_curso=len(retencion)
total_alumnos=len(dp["Correo"].unique())
ratio_1=alumnos_con_mas_de_un_curso/total_alumnos

retencion_2=veces_llevado[veces_llevado>2]
alumnos_con_mas_de_dos_curso=len(retencion_2)
ratio_2=alumnos_con_mas_de_dos_curso/total_alumnos

retencion_3=veces_llevado[veces_llevado>3]
alumnos_con_mas_de_tres_curso=len(retencion_3)
ratio_3=alumnos_con_mas_de_tres_curso/total_alumnos

retencion_4=veces_llevado[veces_llevado>4]
alumnos_con_mas_de_cuatro_curso=len(retencion_4)
ratio_4=alumnos_con_mas_de_cuatro_curso/total_alumnos
print("El ratio para mas de una vez es: ",ratio_1)
print("El ratio para mas de dos veces es: ",ratio_2)
print("El ratio para mas de tres veces es: ",ratio_3)
print("El ratio para mas de cuatro veces es: ",ratio_4)
plt.figure(figsize=(10, 6))
veces_llevado.reset_index().plot()
plt.title('distribucion por veces llevado un curso ')
plt.xlabel('Número de Alumnos')
plt.ylabel('Cantidad de cursos')
plt.show()
```
